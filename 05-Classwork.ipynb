## Bubblegum API

Dear User,

Welcome to the Bubblegum JSON API, available at api.bubblegum.com! Your API key is b335. Endpoints are below.

Note: Your API key must be added to the end of the URL using the parameter APIKEY, like: ?APIKEY=[APIKEY]

### `/flavors/list.json`
Returns a list of the names of all flavors:

```json
{
  "flavors": ["cinnamon", "bubblegum", "wintergreen"]
}
```

---

### `/gum/list.json`
Basic info of every bubble gum:

```json
{
  "gums": [
    { "name": "Minty Fresh", "id": 321 },
    { "name": "Burny", "id": 887 }
  ]
}
```

---

### `/gum/{id}/show.json`
Complete data for a gum:

```json
{
  "id": "321",
  "name": "Minty Fresh",
  "popularity": 200,
  "flavors": ["wintergreen", "mint"]
}
```

---

### `/flavors/{name}.json`
IDs of gums for a given flavor:

```json
{
  "name": "cinnamon",
  "gum_ids": ["321", "887"]
}
```

### How to build URLs

`market` parameter should be `US`

- https://api.bubblegum.com/gum/321/show.json?APIKEY=b335
- https://api.bubblegum.com/gum/321/show.json?APIKEY=b335&market=US&year=2025
- https://api.bubblegum.com/gum/show.json?APIKEY=b335&id=321
- https://api.bubblegum.com?thing=gum&action=show&APIKEY=b335&id=321


### Request a list of all bubble gums, and print the name of each gum.

````python
import requests

url = "https://api.bubblegum.com/gum/list.json?APIKEY=b335"
response = requests.get(url)
data = response.json()


# {
#   "gums": [
#     { "name": "Minty Fresh", "id": 321 },
#     { "name": "Burny", "id": 887 }
#   ]
# }

for gum in data['gums']:
    print(gum['name'])
````

## Paginated APIs

- `page=3` to get the third page of results
- `offset=20` to skip the first 20 results
- by default it gives you 20, `limit=50`, sometimes it might be called `count`
- `?page=3&limit=20` - skips the first two pages of 20, so you get from 41-60
- `?offset=40&limit=20` - skips 40, gives you 41-60

````python
pokemon = ['pikachu', 'eevee', 'snorlax', 'charmander', 'bulbasaur']

for pokemon_name in pokemon:
    response = requests.get(f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}")
    data = response.json()
    
    for stat in data['stats']:
        if stat['stat']['name'] == 'speed':
            speed = stat['base_stat']
            print(data['name'], "has a speed of", speed)
````
